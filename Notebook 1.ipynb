```
import time

# Define win and fail functions
def win():
    print("Congratulations! You have successfully reached Paradise.")
    print("Thanks for playing!")
    exit()

def fail():
    print("You have failed to reach Paradise. Better luck next time.")
    print("Thanks for playing!")
    exit()

# Define game stages
def stage1():
    print("""
    THE DARK FOREST
    """)
    print("""
    You are lost in a dark and terrifying forest, and confronted with the fear and despair that you feel,
    you're wishing for a guide to lead you out of the darkness.

    Suddenly, you see a figure in the distance, dressed in white and holding a light...
    """)

    choice = input("Do you approach the figure or stay put? Enter '1' for approach or '2' for stay: ")

    while choice != "1" and choice != "2" :
        print("Invalid choice. Please enter '1' for approach or '2' for stay.")
        choice = input("Enter '1' for approach or '2' for stay: ")

    if choice == "1":
        print("""You approach the figure with a mixture of fear and hope.
        It is Virgil, the poet and guide sent by Beatrice to lead you through Hell.
        """)
        return 1
    else:
        time.sleep(1)
        print("You decide to stay put and are soon consumed by the despair and suffering of the dark terrifying forest.")
        return 2

def stage2():
    print("""Virgil explains that you must pass through the circles of Hell and depending on the decisions you take, 
you will either meet with Beatrice in Paradise,
or stay with the damned in Hell with punishments fitting to the sins you have committed in life.
    """)
    print("Do you follow Virgil or refuse his guidance?")
    
    choice = input("Enter '1' for 'follow' or '2' for 'refuse': ")
    while choice != "1" and choice != "2" :
        print("Invalid choice. Please enter '1' for 'follow' or '2' for 'refuse'.")
        choice = input("Enter '1' for 'follow' or '2' for 'refuse': ")

    if choice == "1":
        return True
    else:
        return False

def stage3():
    print("You and Virgil begin to make your way through the circles of Hell, encountering various sinners along the way.")
    print("""As you and Virgil continue through the circles of Hell, he begins to explain the concept of 'contrappasso' to you.
         Virgil: 
        'You must understand that the punishment of the souls in Hell is fitting to the sins they committed in life. 
         The punishment corresponds to the nature of the sin and is often a reversal or a perversion of the sin itself.
         This is the law of contrappasso, the suffering is a natural consequence of their actions in life.'
         """)
    print("""
    As you and Virgil pass through Circle VI (Violence), you come across the soul of Julius Caesar.
    He is trapped in a fiery desert, tormented by the memory of his violent actions in life.
    You, moved by Caesar's plight, are now trying to decide if you want to try and help him or let him suffer.
    Do you want to help Julius Caesar by: 
    1. Telling the living of his story and the nature of his punishment, in the hope that it will prevent others from committing the same sin.
    2. Trying to convince him to seek forgiveness and repent for his actions.
    3. Ignore Julius Caesar and let him suffer the pains of the Contrapasso""")
    choice = input("Enter '1', '2' or '3': ")
    while choice != "1" and choice != "2" and choice != "3":
        print("Invalid choice. Please enter '1' or '2'or '3'.")
        choice = input("Enter '1', '2' or '3': ")
    if choice == "1":
        print("You decide to tell the living of Caesar's story and the nature of his punishment.")
        return True
    elif choice == "2":
        print("You try to convince Julius Caesar to seek forgiveness and repent for his actions.")
        return False
    

def stage4():
    print("As you continue on your journey, you come across the entrance to Purgatory.")
    print("Virgil tells you that you have performed enough good deeds to enter, but you must leave him behind.")
    print("Do you enter Purgatory or stay in Hell with Virgil?")
    choice = input("Enter 'enter' or 'stay': ")
    while choice != "enter" and choice != "stay" :
        print("Invalid choice. Please enter 'enter' or 'stay'.")
        choice = input("Enter 'enter' or 'stay': ")
    if choice == "enter":
        return True
    else:
        return False

def stage5():
    print("You find yourself in Purgatory, where souls are cleansed of their sins before reaching Paradise.")
    print("As you make your way through the terraces of Purgatory, you encounter souls who need your help.")
    print("Do you perform good deeds to help them or ignore them?")
    choice = input("Enter 'help' or 'ignore': ")
    while choice != "help" and choice != "ignore" :
        print("Invalid choice. Please enter 'help' or 'ignore'.")
        choice = input("Enter 'help' or 'ignore': ")
    if choice == "help":
        return True
    else:
        return False

def main():
    print("Salve ad Aventum Comoediae Divinae.")
    print("You are about to embark on a journey through Inferno, Purgatorio and Paradiso.")
    print("Your goal is to reach Paradise by performing good deeds along the way.")
    print("You will be presented with various choices and depending on your choices you'll reach paradise or fail.")
    print("""
    You are lost and alone in a dark forest, looking around in fear...
    """)
    print("""
    Oh, how hard it is to tell what that forest was, wild, dense, dark,
    dense, so that the thought of it renews the fear!
    It is so bitter, death is hardly more so; but to speak of the good I found,
    I will tell of the other things I saw.
    """)
    print("""
    I cannot remember well how I entered it, so full was I of sleep when I abandoned the true way;
    but when I had reached the foot of a hill, where the valley that had pierced my heart
    with terror was already such that the firm foot was all of ice, that the heart in the breast doth
    tremble at the recollection, as doth one who, dreaming, believes himself to still be in that affliction
    which he had passed through in the sleep that has just left him.
    """)
    input(prompt="Press enter to begin your journey through Hell!")
    
    # code to control game functioning: proceed to next stage, win and lose
    journey = stage1()
    if journey == 1:
        journey = stage2()
        if journey == 1:
            journey = stage3()
            if journey == 1:
                journey = stage4()
                if journey == True:
                    journey = stage5()
                    if journey == True:
                        win()
                    else: fail()
                else:
                    fail()
            else:
                fail()
        else:
            fail()
    else:
        fail()
main()


```
